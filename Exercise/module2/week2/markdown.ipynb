## Multiplying a vector by a matrix 
* $$\text{Matrix A: } = \space \begin{bmatrix} a_{11} \space ... \space a_{1n} \\\ ... \space ... \space ...\\\ a_{m1} \space ... \space a_{mn} \end{bmatrix}, A \epsilon R^{m*n} $$

* $$\text{Vector: }\space v= \begin{bmatrix} v_1 \\\ v_2 \\\ ... \\\ v_n \end{bmatrix},v \epsilon R^{n}$$
* $$c =Av = \space \begin{bmatrix} a_{11}*v_1 + \space ... \space a_{1n}*n_n \\\ ... \space ... \space ...\\\ a_{m1}*v_1 + \space ... \space a_{mn}* v_n \end{bmatrix} $$

#### Step by Step

```python
np.roots([1,2,-3])  # solution of quadratic equation
print(np.identity(2)) # Identity matrix
```

```python
matrix = np.array([[0.9, 0.2],[0.1, 0.8]])
a, b = matrix[0, 0], matrix[0, 1]
c, d = matrix[1, 0], matrix[1, 1]

trace = a + d
det = a * d - b * c

coefficients = [1, -trace, det]
eigenvalues = np.roots(coefficients) # Solve squaroot equation

print("Matrix A:\n", matrix)
print("Eigenvalues of A:", eigenvalues) 
```

```python
lambda1, lambda2 = eigenvalues
print(lambda1)
print()
print(lambda2)
```

```python
# Solve (A - lambda1*I)v = 0
I = np.identity(2)
a_lambda1_i = matrix - lambda1 * I
print(f'lamda: {a_lambda1_i}')
# Eigenvector corresponding to lambda1
if a_lambda1_i[0, 0] != 0:
    v1_1 = 1
    v1_0 = -a_lambda1_i[0, 1] / a_lambda1_i[0, 0]
else:
    v1_0 = 1
    v1_1 = -a_lambda1_i[1, 0] / a_lambda1_i[1, 1]

eigenvector1 = np.array([v1_0, v1_1])
eigenvector1 = eigenvector1 / np.linalg.norm(eigenvector1)
print(eigenvector1)
```

```python
# Solve (A - lambda2*I)v = 0
a_lambda2_i = matrix - lambda2 * I

if a_lambda2_i[0, 0] != 0:
    v2_1 = 1
    v2_0 = -a_lambda2_i[0, 1] / a_lambda2_i[0, 0]
else:
    v2_0 = 1
    v2_1 = -a_lambda2_i[1, 0] / a_lambda2_i[1, 1]

eigenvector2 = np.array([v2_0, v2_1])
eigenvector2 = eigenvector2 / np.linalg.norm(eigenvector2)

print(eigenvector2)
```

```python
eigenvectors = np.column_stack((eigenvector1, eigenvector2))
print(eigenvectors)
```

## Cosine Similarity 

```python
def cosine_similarity(v1, v2):
    return np.dot(v1,v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

v1 = np.array([1,2,3,4])
v2 = np.array([1,0,3,0])
print(cosine_similarity(v1, v2))
```

## Backgroud Subtraction

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
import cv2
def plot(img):
    img =img.astype(np.uint8)
    rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(rgb, cmap= "gray")
    
new_size = (678,381)
green_background = cv2.imread ("GreenBackground.png",1)    
green_background = cv2.resize(green_background,new_size)

new_background = cv2.imread ("NewBackground.jpg")    
new_background = cv2.resize(new_background,new_size)

object = cv2.imread ("Object.png")  
object = cv2.resize(object,new_size)
```

```python
print(green_background.shape)
print(new_background.shape)
print(object.shape)
```

```python
print(f'B Channel ')
print(object[:,:,0])

print(f'G Channel ')
print(object[:,:,1])

print(f'R Channel ')
print(object[:,:,2])
```

```python
object_copy =object.copy().astype(float)
segment_b  = object_copy[:,:,0]
segment_b [segment_b==40] = 0

segment_g  = object_copy[:,:,1]
segment_g [segment_g==255] = 0

segment_r  = object_copy[:,:,2]
segment_r[segment_r==0] = 0.0

segement  = np.zeros((object_copy.shape), dtype=np.uint8)
segement [:,:,0] = segment_b.astype(float)
segement [:,:,1] = segment_g.astype(float)
segement [:,:,2] = segment_r.astype(float)
segement = segement.astype(np.uint8)

print(segement.shape)
plot(segement)
```

## Create Foregorund Mask

```python
segement_copy = segement.copy().astype(float)
foreground_mask  = np.zeros((object.shape), dtype=np.uint8)


segement_copy[segement_copy[:,:,0]!=0] = 255
foreground_mask[:,:,0] = segement_copy[:,:,0]

segement_copy[segement_copy[:,:,1]!=0] = 255
foreground_mask[:,:,1] = segement_copy[:,:,1]

segement_copy[segement_copy[:,:,2]!=0] = 255
foreground_mask[:,:,2] = segement_copy[:,:,2]

print(foreground_mask.shape)
plot(foreground_mask)
```

```python
foreground_mask = foreground_mask.astype(float)
new_background = new_background.astype(float)

final_ouput = np.zeros((new_background.shape), dtype=np.uint8)

b = (new_background[:,:,0]   -  foreground_mask[:,:,0] )
g = (new_background[:,:,1]   -  foreground_mask[:,:,1] )
r = (new_background[:,:,2]   -  foreground_mask[:,:,2] )

b[b<=0] =0
g[g<=0] =0
r[r<=0] =0

plot(r)
```

```python
final_ouput[:,:,0] = b 
final_ouput[:,:,1] = g
final_ouput[:,:,2] = r

plot(final_ouput + segement)
```
